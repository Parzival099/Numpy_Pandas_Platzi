### Importar librería

```python
import numpy as np 
```

```python
arr = np.linspace(1,10,10, dtype='int8')
arr
```

```python
#Validación elemento a elemento
indices_cond = arr > 5 
indices_cond
```

```python
# Seleccion de elementos basados en alguna condición
arr[indices_cond]
```

```python
#Usando condiciones más complejas
arr[(arr > 5) & (arr<9)]
```

```python
arr[arr > 5] = 99 
```

```python
arr[arr==99]
```
